Wiki Contribution Week 1 

The PH264 course definitely prepared me for this one by giving me an introduction to coding and jupiter notebooks. I hadn't done any coding before so I felt really good about my growth when I finished the course. The core concepts I learned were how to use numpy and matlab, how to analyze large data sets, how to apply mathematical formuli in a computational setting using python, and the methods of chi square minimization. In most physics classes we learn how to describe systems mathematically through the development and testing of models, usually we conduct experiments and use tools like excel to determine if our models accurately define the system and where there is room for improvement. In PH264 we were shown how to use python as a powerful tool to make assessments of our models and how to make them presentable to others, often in less time. For example, I learned that making sure all your arrays are the same size is crucial. I also learned to use the polyfit command in matlab to easily guess unknown values in a model.
