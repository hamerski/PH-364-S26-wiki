### Qutip Probability Distrubtion module ###


For my midterm I have choosen to look at Qutip and more specifically the section that touches upon Probability distribtion of a particle inside a harmonic oscillator. Below will be the link directing to the Qutip website and given library we will use https://qutip.readthedocs.io/en/stable/guide/guide-distributions.html. The given module I have choosen was from qutip.distributions. This module held our analytically solution to the harmonic wave function and harmonic probability density. 

The tutorial I used was the quip.distribution https://nbviewer.org/urls/qutip.org/qutip-tutorials/tutorials-v5/visualization/distributions.ipynb. This allow us to store the harmonic oscillator solution and from there we can use this to plot the real part of the wave function as well as the probability distribution of the harmonic oscillator. 

Qutip is a numerical library that is designed to simulate quantum systems. Some examples of this can be modeling the unitary evolution of a quantum state or expectation value of a given operator. 

In particular our module quip.distribution represents a wave function in position space for a hamiltonian of a given potential. For us this potential would be the harmonic oscillator, while it doesn't actually solve the hamiltonian it can be used to plot the wave function itself in real space and the probability distribution of our wave function. 

```python
from qutip import fock
from qutip.distributions import HarmonicOscillatorProbabilityFunction
import matplotlib.pyplot as plt

M=2
N=20

fig, ax = plt.subplots(M, 1, figsize=(10, 12), sharex=True)

for n in range(M):
    psi = fock(N, n)
    prob = HarmonicOscillatorProbabilityFunction(psi, 1.0, extent=[-10, 10])
    prob.visualize(fig=fig, ax=ax[M-n-1], show_ylabel=False, show_xlabel=(n == 0))
    
fig, ax = plt.subplots(M, 1, figsize=(10, 12), sharex=True)

for n in range(M):
    psi = fock(N, n)
    wf = HarmonicOscillatorWaveFunction(psi, 1.0, extent=[-10, 10])
    wf.visualize(fig=fig, ax=ax[M-n-1], show_ylabel=False, show_xlabel=(n == 0))
```

The code works by first defining the define the dimensions of our Hilbert space in this case we choose N = 20 and M is the number of modes or eigenfunction solutions so in this case M = 2. From there we plot the analytical wave function and probability density for each mode. 

Our first 2 plots are plots are probability distrubtions of a particle in a harmonic oscillator and the last 2 plots are the wave function for our particle in a harmonic oscillator. From the last 2 plots we see how the wave function looks like in real space. The first two plots carry the most information we need. For the second plot we see that it resembles a gaussian function and since the area under the curve tell us probability we see its mostly likely guareenteed to find a particle from roughly a unit distance of -2 to 2. In our first plot we see that there are two distinct spikes in our probability density. This tells us that roughly between unit distance of -2.5 to 0 and 0 to 2.5 we are most likely to find our particle.
